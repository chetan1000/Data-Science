```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_excel('C:/Users/Chetan/Desktop/Data Science Assignments/clustering/EastWestAirlines.xlsx',sheet_name='data')
data.head()
```

```python
#Dropping the columns that have no significance
data1 = data.drop(['ID#','cc1_miles','cc2_miles','cc3_miles','Award?'],axis=1)
data1.head()
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data1_df = scaler.fit_transform(data1)
```

```python
scaled_data1_df
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_data1_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(n_clusters=7, random_state=42)
clusters_new.fit(scaled_data1_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
data1['clusterid_new'] = clusters_new.labels_
data1.head()
```

```python
#these are standardized values.
clusters_new.cluster_centers_
```

```python
data1.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
data1.groupby('clusterid_new').agg(['count']).reset_index()
```
