```python
DF= pd.read_csv("Cars.csv");DF
```

```python
#Stipplot
sns.stripplot(y= 'MPG', data=DF,jitter=True)
plt.ylabel('MPG ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='MPG', y='VOL', data=DF)
plt.ylabel('VOL ($)')
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x='VOL', y='WT', data=DF)
plt.ylabel('WT ($)')
plt.show()
```

```python
#Joint plot

sns.jointplot(x= 'MPG', y= 'VOL', data=DF)
plt.show()
```

```python
# Density plot
sns.jointplot(x='MPG', y= 'SP', data=DF,kind='kde')
plt.show()
```

```python
#pair plot
sns.pairplot(DF)
plt.show()
```
